```python
#Importar librerías
import pandas as pd
import numpy as np
```

```python
#Leemos el contenido del archivo y lo visualizamos
df=pd.read_excel('Inmuebles.xlsx')
#df = pd.read_excel('Inmuebles.xlsx', index_col=4) 
df
```

```python
#Filtro 1, por objeto, cualitativo
filtro1=df[df["Tipo"] == "Parking"]
filtro1
```

```python
#Guardar archivo con el filtro 1 aplicado en csv 
filtro1.to_csv("filtro1.csv")
```

```python
#Filtro 2, por filas no consecutivas, cualitativo
filtro2=df.iloc[ [200,400,600,800,1000,1200,1400,1600,1800,2000,2200,2400,2600,2800,3000,3200] , : ]  
filtro2
```

```python
#Guardar archivo con el filtro 2 aplicado en csv 
filtro2.to_csv("filtro2.csv")
```

```python
#Filtro 3, por columnas consecutivas, cualitativo
filtro3=df.iloc[ : , 6:9]
filtro3
```

```python
#Guardar archivo con el filtro 3 aplicado en csv 
filtro3.to_csv("filtro3.csv")
```

```python
#Filtro 4, de filas vs columnas, cualitativo
filtro4=df.loc[["Barcelona","Girona"], ["Precio Venta"]]
filtro4
```

```python
#Guardar archivo con el filtro 4 aplicado en csv 
filtro4.to_csv("filtro4.csv")
```

```python
#Filtro 5, por cabecera para visualizar las primeras 5 filas de la base de datos, cualitativo
filtro5=df.head()
filtro5
```

```python
#Guardar archivo con el filtro 5 aplicado en csv 
filtro5.to_csv("filtro5.csv")
```

```python
#Filtro 6, por comparación, cuantitativo
filtro6=df[df["Precio Venta"] > 2500000]
filtro6
```

```python
#Guardar archivo con el filtro 6 aplicado en csv 
filtro6.to_csv("filtro6.csv")
```

```python
#Filtro 7, con conector lógico &
filtro7=df[(df["Precio Venta"] > 2000000) & (df["Operación"] == "Alquiler")]
filtro7
```

```python
#Guardar archivo con el filtro 7 aplicado en csv 
filtro7.to_csv("filtro7.csv")
```

```python
#Filtro 8, con conector lógico o
filtro8=df[(df["Superficie"] > 250) | (df["Operación"] == "Venta")]
filtro8
```

```python
#Guardar archivo con el filtro 8 aplicado en csv 
filtro8.to_csv("filtro8.csv")
```
